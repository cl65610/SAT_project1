```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
import statsmodels as sm
import seaborn as sns
%matplotlib inline

# Print the min and max of each column
# The min of rate is below
# print rate_dict
# mins = [x for x in rate_dict if rate_dict[x] == min(rate_dict.itervalues())]
# print 'Minimum states are', [x for x in rate_dict if x in mins], 'with a rate of', min(rate_dict.itervalues())

# To make this easier, I creatd a formula that will accept a dictionary as an
# argument and return the max and min values.

def min_max(dict):
    mins = [x for x in dict if dict[x] == min(dict.itervalues())]
    maxes = [x for x in dict if dict[x] == max(dict.itervalues())]
    if len(mins) > 1:
        print 'The states with the lowest values are: ' , mins
    else:
        print 'The state with the lowest value is: ', mins
    if len(maxes) > 1:
        print 'The states with the highest values are: ', maxes
    else:
        print 'The state with the highest value is: ', maxes

print
print 'For Rate of SATs taken:'
min_max(rate_dict)
print
print 'For SAT Verbal scores:'
min_max(verbal_dict)
print
print 'For SAT Math scores:'
min_max(math_dict)
#Write a function that calculates the standard deviation

def stdev(dict):
    st_dev = np.std([dict[x] for x in dict])
    return st_dev
    
def mean(dict):
    mean = np.mean([dict[x] for x in dict])
    return mean

def median(dict):
    median = np.median([dict[x] for x in dict])
    return median

def mode(dict):
    mode = sp.stats.mode([dict[x] for x in dict])
    return mode

    
# The above formulas will calculate the descriptive statistics required, but the below formula
# will do the same thing from within a single formula.

def desc_stats(dict):
    mean = np.mean([dict[x] for x in dict])
    median = np.median([dict[x] for x in dict])
    mode = sp.stats.mode([dict[x] for x in dict])
    stdev = np.std([dict[x] for x in dict])
    print '''The mean of this data is %.2f.
    The median is: %.2f.
    The mode is %.1f with a count of %d.
    It has a standard deviation of: %.2f''' % (mean, median, mode[0], mode[1], stdev)

print
print 'Rate Statistics'    
desc_stats(rate_dict)
print
print 'Verbal Statistics'
desc_stats(verbal_dict)
print
print 'Math Statistics'
desc_stats(math_dict)

# These list comprehensions create lists to make data visualization below a little bit smoother.
rates = [x[1] for x in data]
states = [x[0] for x in data]
maths = [x[3] for x in data]
verbal =[x[2] for x in data]
```

### Interesting Relationships

As I wrote above, it's a little bit tricky to pull out anything too specific from such limited data. That being said, the data that's present can point us in some interesting directions and help guide our future questions.

The first, most obvious trend we can pull at is that students who tended to do well on on one section of the SAT, also did well on the other section. These data have a Pearson score of 0.9 indicating this strong, positive relationship. 

Unfortunately, this is the only thing that we can say with a great deal of certainty. There are other interesting relationships, but they ask more questions than they provide answers. 
For example, there is a negative relationship between the rate at which students take the SAT and their score on the SAT. The relationship is a bit weaker (Pearson = -0.77), but it still exists and should be investigated further.

The distribution of SAT rate is also interesting and perhaps worthy of further investigation. What is the normal assumption for data distribution?

If data follows a normal distribution, it should behave in predictable ways. Specifically, data points should fall around the mean in specific ways. 95% of all data points should be within two standard deviations of the mean. 65% of all data points should be within one standard deviation. This distribution is what creates the traditional bell-shaped curve when the data is plotted out. 

Does this data follow a normal distribution?

The data for SAT scores in the verbal and math sections seems to follow a normal distribution. However, the rate at which students took the SAT does not appear to follow a normal distribution. It seems closer to a bimodal distrubution.
